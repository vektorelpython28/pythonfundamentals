## try -- except

```python
a = int(input("1. Sayıyı Giriniz:"))
b = int(input("2. Sayıyı Giriniz:"))
print(a/b)
print("Program Çalışmaya Devam Ediyor")
```

```python
try:
    a = int(input("1. Sayıyı Giriniz:"))
    b = int(input("2. Sayıyı Giriniz:"))
    print(a/b)
except:
    print("Hata Verdi")
print("Program Çalışmaya Devam Ediyor")
```

```python
try:
    a = int(input("1. Sayıyı Giriniz:"))
    b = int(input("2. Sayıyı Giriniz:"))
    print(a/b)
except Exception as hata:
    print("Hata Verdi  =>",hata)
print("Program Çalışmaya Devam Ediyor")
```

```python
try:
    a = int(input("1. Sayıyı Giriniz:"))
    b = int(input("2. Sayıyı Giriniz:"))
    print(a/b)
except ValueError:
    print("Değer Hatası")
except ZeroDivisionError:
    print("Sıfıra Bölme Hatası")
print("Program Çalışmaya Devam Ediyor")
```

```python
try:
    a = int(input("1. Sayıyı Giriniz:"))
    b = int(input("2. Sayıyı Giriniz:"))
    print(a/b)
except ValueError:
    print("Değer Hatası")
except Exception:
    print("Genel Hata")
print("Program Çalışmaya Devam Ediyor")
```

### Exception Hierarchy
https://docs.python.org/3/library/exceptions.html

```python
try:
    a = int(input("1. Sayıyı Giriniz:"))
    b = int(input("2. Sayıyı Giriniz:"))
    print(a/b)
except Exception:
    print("Genel Hata")
except ValueError:
    print("Değer Hatası")

print("Program Çalışmaya Devam Ediyor")
```

```python
try:
    a = int(input("1. Sayıyı Giriniz:"))
    b = int(input("2. Sayıyı Giriniz:"))
except ValueError:
    print("Değer Hatası")
else:
    try:
        print(a/b)
    except ZeroDivisionError:
        print("Sıfıra Bölme Hatası")
print("Program Çalışmaya Devam Ediyor")
```
