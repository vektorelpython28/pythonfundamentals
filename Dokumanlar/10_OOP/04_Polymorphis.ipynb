#### Polymorphism

#### Override

```python
class A:
    def __init__(self):
        self.a = "A"

    def soyleA(self):
        return self.a

    def soyle(self):
        print("A Sınıfından Çalıştım")

class B(A):
    def __init__(self): # override
        self.b = "B"


    def soyle(self): #override
        print("B Sınıfından Çalıştım")
    
objA = A()
objB = B()
objB.soyle()
```

#### overload

```python
class A:
    def __init__(self):
        self.a = "A"

    def soyleA(self):
        return self.a

    def soyle(self):
        print("A Sınıfından Çalıştım")

class B(A):
    def __init__(self): # overload
        super().__init__()
        self.b = "B"

    
    def soyle(self): #overload
        super().soyle() #######################################
        print("B Sınıfından Çalıştım")
objA = A()
objB = B()
print(type(objA),"Söyle A fonksiyonu",objA.soyleA())
print(type(objB),"Söyle A fonksiyonu",objB.soyleA())
objB.soyle()
```
