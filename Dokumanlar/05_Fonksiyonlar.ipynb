````python
import random as rnd
from string import ascii_uppercase,ascii_lowercase,digits,punctuation

liste = ascii_lowercase + ascii_uppercase + digits + punctuation
uzunluk = int(input("10 veya 10'dan fazla bir sayı giriniz. "))
def password(a):
    kharf = 0
    bharf = 0
    sayi = 0
    noktalama = 0
    paswd = ""
    for i in range(a):
        indexn = rnd.randint(0,94)
        paswd = paswd + liste[indexn]
        if liste[indexn].islower():
            kharf += 1
        elif liste[indexn].isupper():
            bharf += 1
        elif liste[indexn].isdigit():
            sayi += 1
        else:
            noktalama += 1
    if kharf > 0 and bharf > 0 and sayi>0 and noktalama > 0:
        return paswd
    else:
        password(a)

if uzunluk >= 10:
    print(password(uzunluk))
else:
    print("Hatalı giriş.")
````

````python
import random as rnd
from string import ascii_uppercase,ascii_lowercase,digits,punctuation
def sifreKontrol(sifre):
    bHarf = kHarf = rakam = nokta = False
    for item in sifre:
        if item.isupper():
            bHarf = True
        if item.islower():
            kHarf = True
        if item.isdigit():
            rakam = True
        if item in punctuation:
            nokta = True
    return bHarf and kHarf and rakam and nokta


    
def sifreUret(uzunluk=10):
    if uzunluk >= 10:
        sonuc = ""
        liste = [ascii_uppercase,ascii_lowercase,digits,punctuation]
        while not sifreKontrol(sonuc):
            sonuc = ""
            for i in range(uzunluk):
                sonuc += rnd.choice(rnd.choice(liste))
        else:
            return sonuc

    else:
        return "10 ya da daha büyük sayı giriniz"
sifreUret(12)
````

Girilen bir sayının Armstrong Sayısı olup olmadağını tespit eden bir python fonksiyonu yazınız


````python

"""
371 => 3 ** 3 + 7 ** 3 + 1 ** 3 => 371 == 371
"""

def armstrong(sayi):
    sayi = str(sayi)
    uzunluk = len(sayi)
    toplam = 0
    for item in sayi:
        toplam += int(item)**uzunluk
    return toplam == int(sayi)

armstrong(372)
````

````python
armstrong = lambda sayi : sayi == sum([int(item) ** len(list(str(sayi))) for item in list(str(sayi))])
armstrong(371)
for i in range(1,1000):
    if armstrong(i):
        print(i)
````

````python
def is_defended(attackers, defenders):
    deff = att = 0
    uzun,elem = len(attackers) if len(attackers) >= len(defenders) else  len(defenders)
    for i in range(uzun):
        if attackers[i:i+1] > defenders[i:i+1]:
            att += 1
        else:
            deff += 1
    if deff == att:
        return sum(attackers) <= sum(defenders)
    else:
        return deff > att
````
